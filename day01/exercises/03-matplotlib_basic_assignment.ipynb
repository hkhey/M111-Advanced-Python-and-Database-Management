## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
# Write your code here
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\acer\Desktop\M111-Advanced-Python-and-Database-Management\day01\data\company_sales_data.csv")
X = data['month_number']
Y = data['total_profit']
plt.plot(X, Y)
plt.ylabel('Profit in dollar')
plt.xlabel("Month number")
plt.title("Company profit per month")
```

## Exercise 4

## Exercise 7
**Read the total profit of each month and show it using the histogram to see the most common profit ranges**

**Expected output**

<img src="../data/Profit_data.png" alt="Profit_data" style="width: 700px;"/>


```python
# write your code here
range_profit = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(data['total_profit'], range_profit, label = 'Profit data')
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(range_profit)
plt.title('Profit data')
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
# write your code here
data2   = [data ['bathingsoap'].sum(), data ['facewash'].sum(), data ['facecream'].sum(), data ['toothpaste'].sum(), data ['shampoo'].sum(), data ['moisturizer'].sum()]
fig = plt.figure(figsize =(7, 9))

plt.title('Sales data')
plt.axis("equal")
plt.pie(data2, labels = ['Bathing soap', 'FaseWash', 'FaceCream', 'ToothPaste', 'Shampoo', 'Moisturizer'], autopct='%1.1f%%')
plt.legend(loc='lower right')
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
# write your code here
fig = plt.figure(figsize =(7, 9))

plt.subplot(2, 1, 1)
plt.plot(X, data['bathingsoap'], color='black', marker='o', linewidth=3)
plt.title('Sales data of  a Bathingsoap')

plt.subplot(2, 1, 2)
plt.plot(X, data['facewash'], color='red', marker='o', linewidth=3)
plt.title('Sales data of  a facewash')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
# write your code here
fig = plt.figure(figsize =(7, 7))

labels = ['FaceCream', 'FaseWash', 'ToothPaste','Bathing soap', 'Shampoo', 'Moisturizer']
plt.stackplot(X, data['facecream'], data['facewash'], data['toothpaste'], data['bathingsoap'], data['shampoo'], data['moisturizer'], labels=labels)
plt.legend(loc = 'upper left')
plt.title('Alll product sales data using stack plot')
plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
```
